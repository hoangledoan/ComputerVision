### Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

# Read in the image
image = cv2.imread('D:/ComputerVision/images/thumbs_up_down.jpg')

# Change color to RGB (from BGR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

### Produce a binary image for finding contours

```python
# Convert to gray scale
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# Creat a binary thresholed image
# cv2.THRESH_BINARY_INV set all the white scale from 225-255 to black
retval, binary = cv2.threshold(gray, 225, 255, cv2.THRESH_BINARY_INV)

plt.imshow(binary, cmap='gray')
```

### Find and draw the contours

```python
# Find contours from thresholded, binary image
# cv2.RETR_TREE gives the order of objects which are nested in each other(hierachy)
# cv2.CHAIN_APPROX_SIMPLE compress horizontal, vertical and diagonal segments and leaves their end points.
# for example: rectangular returns 4 points
contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

contours_image = np.copy(image)

# Draw the contour on the image
# -1: full pixel
contours_image = cv2.drawContours(contours_image, contours, -1, (0,255,0), 3)
plt.imshow(contours_image)
```
